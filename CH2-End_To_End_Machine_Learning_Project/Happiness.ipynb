```python
import pandas as pd
import os

csv_path = os.path.join("datasets", "happiness", "world_happiness_report.csv")
happiness = pd.read_csv(csv_path)
```

```python
happiness.head()
```

```python
happiness.info()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

corr_matrix = happiness.corr(numeric_only=True)
corr_matrix["Happiness_Score"].sort_values(ascending = False)
```

```python
from pandas.plotting import scatter_matrix

attributes = ["Happiness_Score", "Freedom", "Unemployment_Rate", "Income_Inequality"]
scatter_matrix(happiness[attributes], figsize=(12,8))
```
